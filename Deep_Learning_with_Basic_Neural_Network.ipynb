# Create Neural Network with relu activation function

```python
#diagram for which we have to solve our neural network
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Path to your image file
image_path = '/home/abhishek/Downloads/02.jpg'

# Load and display the image
img = mpimg.imread(image_path)
plt.imshow(img)
plt.axis('off')  # Hide axis
plt.show()
```

```python
def relu(input):
    '''Define your relu activation function here'''
    # Calculate the value for the output of the relu function: output
    output = max(0, input)

    # Return the value just calculated
    return(output)
```

```python
relu(9)
```

```python
relu(-9)
```

```python
# Calculate node 0 value: node_0_output
node_0_input = (input_data * weights['node_0']).sum()
print(node_0_input)
node_0_output = relu(node_0_input)
print(node_0_output)
```

```python
# Calculate node 1 value: node_1_output
node_1_input = (input_data * weights['node_1']).sum()
print(node_1_input)
node_1_output = relu(node_1_input)
print(node_1_output)
```

```python
# Put node values into array: hidden_layer_outputs
hidden_layer_outputs = np.array([node_0_output, node_1_output])
print(hidden_layer_outputs)
```

```python
# Calculate model output (do not apply relu)
model_output = (hidden_layer_outputs * weights['output']).sum()
```

```python
# Print model output
print(model_output)
```

# create neural network with two hidden layer

```python
#diagram for which we have to solve our neural network
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Path to your image file
image_path = '/home/abhishek/Downloads/03.jpg'

# Load and display the image
img = mpimg.imread(image_path)
plt.imshow(img)
plt.axis('off')  # Hide axis
plt.show()
```

```python
input_data=np.array([3,5])
```

```python
weights={'node_0_0':np.array([2,4]),
         'node_0_1':np.array([4,-5]),
         'node_1_0':np.array([0,1]),
         'node_1_1':np.array([1,1]),
         'output':np.array([2,7])}
```

```python
weights
```

```python
# Calculate node_0_0 in the first hidden layer
node_0_0_input = (input_data * weights['node_0_0']).sum()
print(node_0_0_input)
node_0_0_output = relu(node_0_0_input)
print(node_0_0_output)
```

```python
# Calculate node_0_1 in the first hidden layer
node_0_1_input = (input_data * weights['node_0_1']).sum()
print(node_0_1_input)
node_0_1_output = relu(node_0_1_input)
print(node_0_1_output)
```

```python
  # Put node values into array: hidden_0_outputs
hidden_0_outputs = np.array([node_0_0_output, node_0_1_output])
hidden_0_outputs
```

```python
# Calculate node_1_0 in the second hidden layer
node_1_0_input = (hidden_0_outputs * weights['node_1_0']).sum()
print(node_1_0_input)
node_1_0_output = relu(node_1_0_input)
print(node_1_0_output)
```

```python
# Calculate node_1_1 in the second hidden layer
node_1_1_input = (hidden_0_outputs * weights['node_1_1']).sum()
print(node_1_1_input)
node_1_1_output = relu(node_1_1_input)
print(node_1_1_output)
```

```python
# Put node values into array: hidden_1_outputs
hidden_1_outputs = np.array([node_1_0_output, node_1_1_output])
hidden_1_outputs
```

```python
# Calculate output here: model_output
model_output = (hidden_1_outputs * weights['output']).sum()
print(model_output)
```
